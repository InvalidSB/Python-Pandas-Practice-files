# Daily percentage change


```python
tesla['returns']=(tesla['Close']/tesla['Close'].shift(1))-1
```

```python
tesla.head()
```

```python
tesla['returns'].hist(bins=100)
```

```python
ford['returns']=(ford['Close']/ford['Close'].shift(1))-1
gm['returns']=(gm['Close']/gm['Close'].shift(1))-1
tata['returns']=(tata['Close']/tata['Close'].shift(1))-1
```

```python
figure(figsize=(15, 9), dpi=300)
ford['returns'].hist(bins=100,label="ford",alpha=0.7)
gm['returns'].hist(bins=100,label="gm",alpha=0.7)
tata['returns'].hist(bins=100,label="tata",alpha=0.7)
tesla['returns'].hist(bins=100,label="tasla",alpha=0.3,color="k")
# more flatter it is more stable it will be
plt.legend()
plt.show()
```

```python
figure(figsize=(15,12))
tesla['returns'].plot(kind="kde",label="tesla")
ford['returns'].plot(kind="kde",label="ford")
tata['returns'].plot(kind="kde",label="tata")
gm['returns'].plot(kind="kde",label="gm")
plt.legend()
plt.show()
# pdf so it is normalized
```

```python
box_df=pd.concat([tesla['returns'],gm['returns'],ford['returns'],tata['returns']],axis=1)
box_df.columns=(['Tesla',"GM","Ford","Tata"])
box_df.plot(kind='box')
plt.show()
```

# cumilative returns

```python
tesla["cumilativeReturns"]=(1+ tesla['returns']).cumprod()
tesla.head()
```

```python
tata["cumilativeReturns"]=(1+ tata['returns']).cumprod()
ford["cumilativeReturns"]=(1+ ford['returns']).cumprod()
gm["cumilativeReturns"]=(1+ gm['returns']).cumprod()
```

```python
figure(figsize=(15, 9), dpi=300)
tesla['cumilativeReturns'].plot(label="tesla")
ford['cumilativeReturns'].plot(label="ford")
gm['cumilativeReturns'].plot(label="gm")

tata['cumilativeReturns'].plot(label="tata")
plt.legend()
plt.show()
```
